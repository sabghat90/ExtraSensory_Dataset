### Outlier detection

```python
from scipy import stats
from scipy.stats import zscore
from sklearn.ensemble import RandomForestClassifier
```

```python
# Drop non-numeric columns for outlier detection
numeric_df = df.drop(columns=['diagnosis'])
```

```python
# Box Plot for Outlier Detection
plt.figure(figsize=(20, 15))
sns.boxplot(data=numeric_df)
plt.title('Box Plot for Outlier Detection')
plt.xticks(rotation=90)
plt.show()
```

```python
# Z-Score Method for Outlier Detection
z_scores = np.abs(zscore(numeric_df))
outliers_z = (z_scores > 3).any(axis=1)
print(f"Number of outliers detected using Z-Score: {outliers_z.sum()}")
```

```python
# Display rows with outliers (Z-Score)
outliers_z_df = df[outliers_z]
print("Outliers detected using Z-Score:")
print(outliers_z_df)
```

```python
# IQR Method for Outlier Detection
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1
outliers_iqr = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).any(axis=1)
print(f"Number of outliers detected using IQR: {outliers_iqr.sum()}")
```

```python
# Display rows with outliers (IQR)
outliers_iqr_df = df[outliers_iqr]
print("Outliers detected using IQR:")
print(outliers_iqr_df)
```

```python
from sklearn.ensemble import IsolationForest

# Isolation Forest
iso_forest = IsolationForest(contamination=0.05, random_state=42)
outliers_iso = iso_forest.fit_predict(numeric_df) == -1
print(f"Number of outliers detected using Isolation Forest: {outliers_iso.sum()}")
```

```python
# Remove outliers
df_cleaned = df[~outliers_iqr]
print(f"Dataset size after removing IQR outliers: {df_cleaned.shape}")
```
